```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pokemon_data=pd.read_csv("pokemon.csv")
pokemon_data.columns
# print(pokemon_data)
```

```python
# pokemon_data=pd.read_csv("5_Matplotlib_and_Seaborn/pokemon.csv")

# sb.countplot(data=pokemon_data,x="Generation")
# creates a bar graph 
sb.countplot(data=pokemon_data,x="Generation");
```

```python
sb.countplot(data=pokemon_data,x="Generation",color="red");
# sb.color_palette() ---> give the list of all colors available in seaborn
sb.color_palette()
```

```python
# shows the graph in particular order
sb.countplot(data=pokemon_data,x="Generation",color="red" ,order=[1,4,2,3,6,5]);
```

```python
# value_counts method Return a Series containing counts of unique values 
descending_gen=pokemon_data["Generation"].value_counts().index
print(descending_gen)
sb.countplot(data=pokemon_data,x="Generation",color="red" ,order=descending_gen);
```

```python
sb.countplot(data=pokemon_data,y="Type 1",color="red",order=pokemon_data["Type 1"].value_counts().index);
```
