```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def showDataFrame(DataFrame,hideIndex=False):
    if hideIndex == True:
        print(DataFrame.to_string(index=False))
        return
    print(DataFrame.to_markdown())
    print()
```

# Distplot
distribtion plot \
it shows data distribution of data against density distribution

```python
sb.distplot([1,2,3,4,5]);
```

```python
data=np.random.random(20)
sb.distplot(data);
```

```python
data=np.random.rand(20) #between 0-1
data=np.random.random(20) #between -1 and 1 
# hides the histogram in the background
sb.distplot(data,hist=0);
```

# Line Plots

```python
x=np.random.randint(0,11,5)
y=np.random.randint(100,200,5)
temp_dict={"Number":x,"Cost":y}
df=pd.DataFrame(temp_dict,index=range(1,len(x)+1))
sb.lineplot(x="Number",y="Cost",data=df)
```

```python
# line plot using matplotlib
plt.plot(x,y)
```

```python
irisData=sb.load_dataset('iris') # inbuilt data set
sb.lineplot(x="sepal_length",y="sepal_width",data=irisData);
```

```python
tipsData=sb.load_dataset("tips")
# hue  groups the variable and shows different lines for all the unique values in the given column 
sb.lineplot(x="total_bill",y="tip",hue="sex",style="sex",data=tipsData)
```

# LMplot
it shows a line representing a linear regression model

```python
df=sb.load_dataset("anscombe")
showDataFrame(df)
sb.lmplot(x="x",y="y",data=df)
```

```python
sb.lmplot(x="total_bill",y="tip",data=tipsData)
```
