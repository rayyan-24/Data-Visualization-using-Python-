```python
def showDataFrame(DataFrame,hideIndex=False):
    if hideIndex == True:
        print(DataFrame.to_string(index=False))
        return
    print(DataFrame.to_markdown())
    print()
```

### Using basic attributes of plot function 

```python
x=[1,2,3]
y=[10,4,19]
'''
points of plotting will be A(1,10) B(2,4) B(3,19)
at plotting points A B C a marker will be shown
color gives the color to line
label gives a name
'''
plt.plot(x,y,marker='o',color="g",linewidth=1,label="LINE 1")
plt.legend()
```

# Bar charts
## Vertical Bar charts
bar charts are used for categorical data \n
plt.bar()

```python
subjects=["English","Hindi","Urdu","Maths"]
marks=[98,78,88,89]
'''
x is the data to be plotted on x axis
height is the values of y axis
width is the width of each bar
'''
plt.bar(x=subjects,height=marks,color="b",width=0.1);
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Wise Marks Distribution")
```

# Horizontal Bar Plots

```python
subjects=["English","Hindi","Urdu","Maths"]
marks=[98,78,88,89]

'''
y is the data to be plotted on y axis
width is the data for x axis
height is the width of each bar
'''

plt.barh(y=subjects,width=marks,height=0.1,color="g")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Wise Marks Distribution");
```

# Pie Charts

```python
# created in anti-clockwise direction
data=[20,56,64,69,25]
labels=["AUDI","BENZ","TESLA","BMW","FORD"]
plt.pie(data,labels=labels)
plt.show();
```

```python
'''
wedges are the different colored areas slices
By Default First wedge starts at 0 deg as measurd through protactor (positive x axis)
explode param takes a tuple which shows the values to be highlighted
autopct shows the format of percentages to be shown on the wedges
startangle gives the start angle
'''
data=[20,56,64,69,25]
labels=["AUDI","BENZ","TESLA","BMW","FORD"]
explode=(0,0,0.1,0,0)
plt.pie(x=data,explode=explode,labels=labels,autopct='%.1f%%',shadow=0,startangle=90)
plt.show();
```

```python
data=[20,56,64,69,25]
labels=["AUDI","BENZ","TESLA","BMW","FORD"]
explode=(0,0,0.1,0,0)
colors=["red","green","blue","pink","black"]
plt.pie(x=data,explode=explode,labels=labels,autopct='%.1f%%',shadow=0,startangle=90,colors=colors)
plt.show();
```

# Area Plot
area plot is an extention of line chart it is obtained by filling color between line and the axis(x axis)

```python
'''
alpha in line graph depicts the transparency of line range lies between 0 and 1 both inclusive
like rgba(0-255,0-255,0-255,0-1) in CSS
'''

x=range(1,6)
y=[1,4,8,8,4]
plt.plot(x,y,marker="o",color="blue",alpha=1) # line blue
plt.show();
plt.fill_between(x,y,color="green"); # colored region green
plt.show();
```

```python
x=range(1,6)
y=[1,4,8,8,4]
plt.plot(x,y,marker="o",color="blue",alpha=0.9) # line blue
plt.fill_between(x,y,color="green",alpha=0.7); # colored region green
plt.show();
```

# Stacked Area Plot

```python
'''
y parameter can be passed as list of list or indivisual lists
loc in plt.legend() means the location of legend
'''
x=range(1,6)
y=[[1,4,6,8,9],[2,2,7,10,12],[2,8,5,10,6]]
label=["A","B","C"]
col=["red","blue","green"]
plt.stackplot(x,y,labels=label,colors=col)
plt.legend(loc="upper left")
plt.show()
```

```python
x=range(1,6)
y=[[1,4,6,8,9],[2,2,7,10,12],[2,8,5,10,6]]
label=["A","B","C"]
col=["red","blue","green"]
plt.stackplot(x,y[0],y[1],y[2],labels=label,colors=col)
plt.legend(loc="upper left")
plt.show()
```
